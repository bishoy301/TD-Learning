```python
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sampleOne():
    return np.random.normal(3, 1) + 1j * np.random.normal(-1, 1)

def sampleZero():
    return np.random.normal(-1, 1) + 1j * np.random.normal(1, 2)
```

```python
Y_new = []
Y_test = []
X1_train = np.array([(c.real, c.imag) for c in map(lambda _: sampleOne(), range(100))])
X0_train = np.array([(c.real, c.imag) for c in map(lambda _: sampleOne(), range(100))])

X1_test = np.array([(c.real, c.imag) for c in map(lambda _: sampleOne(), range(100))])
X0_test = np.array([(c.real, c.imag) for c in map(lambda _: sampleOne(), range(100))])

X_new = X1_train
X_new = np.concatenate((X_new, X0_train))

for i in range(100):
    Y_new.append(1)
for i in range(100):
    Y_new.append(0)

Y_new = np.array(Y_new)

c = list(zip(X_new, Y_new))
np.random.shuffle(c)
X_new, Y_new = zip(*c)
X_new = np.array(X_new)
Y_new = np.array(Y_new)
```

```python
n_inputs = 2
n_hidden = 2
n_outputs = 2

X = tf.placeholder(tf.float32, shape=(None, n_inputs), name='X')
Y = tf.placeholder(tf.int64, shape=(None), name='Y')
```

```python
def layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        stddev = 2 / np.sqrt(n_inputs)
        # using truncated normal instead of regular normal so that there are no weights that are too large and slow down training
        init = tf.truncated_normal((n_inputs, n_neurons), stddev=stddev) 
        w = tf.Variable(init, name='weights')
        b = tf.Variable(tf.zeros([n_neurons]), name='biases')
        z = tf.matmul(X, w) + b
        
        if activation == 'relu':
            return tf.nn.relu(z)
        else:
            return z
```

```python
with tf.name_scope('NN'):
    hidden = layer(X, n_hidden, 'hidden', activation='relu')
    hidden2 = layer(hidden, n_hidden, 'hidden2', activation='relu')
    logits = layer(hidden, n_outputs, 'outputs')
```

```python
with tf.name_scope('loss'):
    cross_entropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=Y, logits=logits)
    loss = tf.reduce_mean(cross_entropy, name='loss')
```

```python
learning_rate = 0.001
with tf.name_scope('train'):
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
    training_op = optimizer.minimize(loss)
```

```python
with tf.name_scope('eval'):
    correct = tf.nn.in_top_k(logits, Y, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
n_epochs = 2000
batch_size = 16

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        total_batch = int(len(X_new)/batch_size)
        X_batches = np.array_split(X_new, total_batch)
        Y_batches = np.array_split(Y_new, total_batch)
        for i in range(total_batch):
            batch_x, batch_y = X_batches[i], Y_batches[i]
            sess.run(training_op, feed_dict={X: batch_x, Y: batch_y})
        acc_train = accuracy.eval(feed_dict={X: batch_x, Y: batch_y})
        print(epoch, "Train Accuracy:", acc_train)
```
